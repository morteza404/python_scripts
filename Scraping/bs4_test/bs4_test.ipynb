```python
from bs4 import BeautifulSoup
import urllib
```

```python
url = "https://www.pythonforbeginners.com"

content = urllib.request.urlopen(url).read()

soup = BeautifulSoup(content)

print(soup.prettify())
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
print(soup.get_text())
```
